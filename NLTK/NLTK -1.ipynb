```python
class text:
    def __init__(self,file):
        self.file = file
    def sentToken(self,file):
        self.sentence = nltk.tokenize.sent_tokenize(self.file)
        return self.sentence
    def wordToken(self,file):
        self.words = nltk.tokenize.word_tokenize(self.file)
        return self.words
    def puncktToken(self,file):
        self.pwords=nltk.tokenize.PunktSentenceTokenizer(self.file)
        return self.pwords
    def stem(self,var):
        ps=nltk.stem.PorterStemmer()
        self.stem = [ps.stem(word) for word in var]
        return self.stem
    def stop(self,var,stop_words):
        self.clean = [w for w in var if w not in stop_words]
        return self.clean
    def pos(self,var):
        self.pos=nltk.pos_tag(var)
        return self.pos
```

### Words as Features
Now we will try to convert the words in features for this we have following steps.
<br> 1. get the top most no of words from all_words (word_features)
<br> 2. then write a function that gets the document and gives us the words with boolen true or false.

```python
word_features = list(all_words.keys())[:3000]
```

```python
def find_features(document):
    word=set(document)
    feature={}
    for w in word_features:
        feature[w]=(w in word)
    return feature
        
```

```python
print(find_features(movie_reviews.words('neg/cv000_29416.txt')))
```

## Sentiment Analysis module

```python
short_pos = open('C:/Users/Top spot/Dropbox/NLTK/short_reviews/positive.txt','r').read()
short_neg = open('C:/Users/Top spot/Dropbox/NLTK/short_reviews/negative.txt','r').read()
documents = []
all_words=[]
all_allowed_wordtype=['J','R','V']
#all_allowed_wordtype=['J']
for r in short_pos.split('\n'):
    documents.append((r,'pos'))
    words=word_tokenize(r)
    pos=nltk.pos_tag(words)
    for w in pos:
        if w[1][0] in all_allowed_wordtype:
            all_words.append(w[0].lower())
for r in short_neg.split('\n'):
    documents.append((r,'neg'))
    words=word_tokenize(r)
    neg=nltk.pos_tag(words)
    for w in neg:
        if w[1][0] in all_allowed_wordtype:
            all_words.append(w[0].lower())
```

```python
all_words=nltk.FreqDist(all_words)
```

```python
features=list(all_words.keys())[:5000]
```

```python
def find_features(document):
    words = set(word_tokenize(document)) # as documents are sentences here so we have to make them words before comparision
    feature={}
    for w in features:
        feature[w]=(w in words)
    return feature
```

```python
featuresets=[(find_features(rev),cat) for (rev,cat) in documents]
```

```python
random.shuffle(featuresets)
```

```python
training=featuresets[:10000]
test = featuresets[10000:]
```

```python
classifier = NaiveBayesClassifier.train(training)
print("Original classifier accuracy: ",nltk.classify.accuracy(classifier,test))
```

```python
MultinomialNB_classifier = SklearnClassifier(MultinomialNB())
MultinomialNB_classifier.train(training)
print("MultinomialNB classifier accuracy: ",nltk.classify.accuracy(MultinomialNB_classifier,test))
```
